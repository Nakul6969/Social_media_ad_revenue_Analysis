```python
import pandas as pd

df=pd.read_excel("social_media_ad_revenue_15k.xlsx")
```

```python
print(df.shape)
```

```python
print(df.columns)
```

```python
print(df.isnull().sum())
```

```python

df['Date'] = pd.to_datetime(df['Date'])
df['Date'].iloc[0] = df['Date'].iloc[1] - pd.Timedelta(days=1)
```

```python
df = df.dropna(subset = ['Platform'])
```

```python
df['Ad_Spend'] = df['Ad_Spend'].fillna(df['Ad_Spend'].mean())
```

```python
df['Clicks']=df['Clicks'].fillna(0)
df['Impressions']=df['Impressions'].fillna(0)
df['Conversions']=df['Conversions'].fillna(0)
```

```python
df['Revenue'] = df.groupby('Platform')['Revenue'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
# CPC = Ad_Spend / Clicks  { Cost Per Click }

df['CPC'] = df.apply(lambda x: x['Ad_Spend'] / x['Clicks'] if x['Clicks'] > 0 else 0, axis=1)
```

```python
# CPM = (Ad_Spend / Impressions)*1000 { Cost per Thousand }
df['CPM'] = df.apply(lambda x: (x['Ad_Spend'] / x['Impressions']) * 1000
                     if x['Impressions'] > 0 else 0, axis=1)
```

```python
# CTR = Clicks / Impressions { Cost Through Rate}

df['CTR'] = df.apply(lambda x: (x['Clicks'] / x['Impressions']) 
                     if x['Impressions'] > 0 else 0, axis=1)
```

```python
print(df.shape)

print(df.isnull().sum())
```

```python
df.to_excel("cleaned_dataset.xlsx", index=False)
```
